```python
import numpy as np
from wordcloud import WordCloud
from PIL import Image
import matplotlib.pyplot as plt
```

```python
photo = "galloping-horse-vector.jpg"
```

```python
icon = Image.open(photo)

plt.figure(figsize = (10,5))
plt.imshow(icon)
plt.axis('off')
plt.show()
```

```python
ex_file = "SpamData2/01_Processing/practice_email.txt"
```

```python
stream = open(ex_file, encoding= 'latin-1')

is_body = False
lines = []

for line in stream:
    if is_body:
        lines.append(line)
        
    elif line == '\n':
        is_body = True
        
stream.close()
```

```python
lines
```

```python
email = ''.join(lines)
```

```python
print(email)
```

```python
image_mask_1 = Image.new(mode = 'RGB', size = icon.size, color=(0,0,0))
image_mask_1.paste(icon, box = None)

rgb_array = np.array(image_mask_1)

word_cloud = WordCloud(mask = rgb_array)

word_cloud.generate(email)
plt.figure(figsize =(18,16))

plt.imshow(word_cloud, interpolation= 'bilinear')

plt.axis('off')
plt.show()
```

```python
image_mask_1 = Image.new(mode = 'RGB', size = icon.size, color=(255,255,255))
image_mask_1.paste(icon, box = None)

rgb_array = np.array(image_mask_1)

word_cloud = WordCloud(mask = rgb_array, background_color= 'white', max_words= 6000, colormap= 'plasma', repeat= True)

word_cloud.generate(email)
plt.figure(figsize =(18,16))

plt.imshow(word_cloud, interpolation= 'bilinear')

plt.axis('off')
plt.show()
```

```python
icon = Image.open(photo)
image_mask_1 = Image.new(mode = 'RGB', size = icon.size, color=(255,255,255))
image_mask_1.paste(icon, box = None)

rgb_array = np.array(image_mask_1)

word_cloud = WordCloud(mask = rgb_array, background_color= 'black', max_words= 30000, colormap= 'plasma', repeat= True)

word_cloud.generate(email)
plt.figure(figsize =(18,16))

plt.imshow(word_cloud, interpolation= 'bilinear')

plt.axis('off')
plt.show()
```
