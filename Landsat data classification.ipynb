```python
import pandas as pd

with open('sat.trn') as f:
    data = f.readlines()

training_data =pd.DataFrame([list(map(int, line.split())) for line in data])

training_data.columns = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z',
               'aa','ab','ac','ad','ae','af','ag','ah','ai','aj','class']

training_data
```

```python
with open('sat.tst') as f:
    data = f.readlines()

test_data =pd.DataFrame([list(map(int, line.split())) for line in data])

test_data.columns = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z',
               'aa','ab','ac','ad','ae','af','ag','ah','ai','aj','class']

test_data
```

```python
x_train = training_data.iloc[:, :-1]

y_train = training_data.iloc[:,-1]
```

```python
x_test = test_data.iloc[:, :-1]

y_test = test_data.iloc[:,-1]
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()

model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
```

```python
predictions
```

```python
from sklearn.metrics import accuracy_score

score = accuracy_score(y_test, predictions)

score
```

```python
pred_df = pd.DataFrame({'': test_data.iloc[:, 0], '': predictions})

pred_df
```

```python
pred_df.to_csv('Task3-predictions.csv', header=False, index=False)
```
