# Missing values

```python
feature_with_nan = [features for features in dataset.columns if dataset[features].isnull().sum()>1]
```

```python
for feature in feature_with_nan:
    print(feature,np.round(dataset[feature].isnull().mean(), 4), '% missing values')
```

### There are many missing values , we need to find the relation between the missing values and the sales price

```python
for feature in feature_with_nan:
    data = dataset.copy()
    
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

# Numerical variables

```python
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtype != 'O']   
print('Length of numerical feature:',len(numerical_features))

dataset[numerical_features].head()
```

# Temporal Datetime variable

```python
year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]

year_feature
```

```python
for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Yr_sold')
plt.ylabel('Sale_price')

plt.title('House Price Vs Yr_sold')
```

```python
for feature in year_feature:
    if feature != 'YrSold':
        data = dataset.copy()
        
        data[feature] = data['YrSold']-data[feature]
        
#        plt.figure(figsize =(5,4))
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
# Numerical variables are basically 2 types
# 1 => continuous variables      2 => Descrete variables

descrete_feature = [feature for feature in numerical_features if len(dataset[feature].unique()) < 25 and feature not in year_feature + ['Id']]
print('Descrete variable count: {}'.format(len(descrete_feature)))
```

```python
descrete_feature
```

```python
dataset[descrete_feature].head()
```

```python
# Find the relation between descrete_feature and SalePrice

for feature in descrete_feature:
    data =dataset.copy()
    
    data.groupby(feature)['SalePrice'].median().plot.bar()
    
#    plt.figure(figsize =(5,4))
    plt.xlabel(feature)
    plt.ylabel('Sale_price')
    plt.title(feature)
    plt.show()
```

# Continuous variable

```python
continuous_feature = [feature for feature in numerical_features if feature not in descrete_feature + year_feature + ['Id']]
print('Continuous feature count {}'.format(len(continuous_feature)))
```

```python
for feature in continuous_feature:
    data = dataset.copy()
    
#    plt.figure(figsize =(5,4))
    data[feature].hist(bins = 25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

```python
# logirothemic transormation

for feature in continuous_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        
#        plt.figure(figsize=(5,4))
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

# Outliers

```python
for feature in continuous_feature:
    data = dataset.copy()
    
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        
        data.boxplot(column = feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
