```python
from wordcloud import WordCloud
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
Example_file = 'SpamData2/01_Processing/practice_email.txt'
```

```python
stream = open(Example_file, encoding = 'latin-1')

is_body = False

lines = []

for line in stream:
    if is_body:
        lines.append(line)
    
    elif line == '\n':
        is_body = True
    
stream.close()
```

```python
lines
```

```python
email_body = ''.join(lines)
```

## WordCloud

```python
plt.figure(figsize=(18,8))

word_cloud = WordCloud().generate('one one one two two  three three three three four four four four four four four five')

plt.imshow(word_cloud)
plt.show()
```

WordCloud shows the words in dessending order...... most usage word is shown at a larger size.... least usage word shown at a smaller size

```python
whale_file = 'SpamData/01_Processing/wordcloud_resources/whale-icon.png'
```

```python
icon = Image.open(whale_file)

plt.figure(figsize = (10,5))
plt.imshow(icon)
plt.axis('off')
plt.show()
```

```python
image_mask = Image.new(mode ='RGB', size = (24,16), color =(255,255,255))

plt.imshow(image_mask)
plt.show()
```

```python
image_mask = Image.new(mode ='RGB', size = (24,16), color =(155,50,150))

plt.imshow(image_mask)
plt.show()
```

```python
image_mask_1 = Image.new(mode ='RGB', size =icon.size, color =(255,255,255))       # creating a white background
image_mask_1.paste(icon, box = icon)

rgb_array = np.array(image_mask_1)
word_cloud = WordCloud(mask = rgb_array)
word_cloud.generate(email_body)

plt.imshow(word_cloud, interpolation= 'bilinear')
plt.axis('off')
plt.show()
```

```python
image_mask_1 = Image.new(mode ='RGB', size =icon.size, color =(255,255,255))
image_mask_1.paste(icon, box = icon)

plt.figure(figsize=(18,8))
rgb_array = np.array(image_mask_1)
word_cloud = WordCloud(mask = rgb_array)
word_cloud.generate(email_body)

plt.imshow(word_cloud, interpolation= 'bilinear')
plt.axis('off')
plt.show()
```

```python
image_mask_1 = Image.new(mode ='RGB', size =icon.size, color =(255,255,255))
image_mask_1.paste(icon, box = icon)

plt.figure(figsize=(18,8))
rgb_array = np.array(image_mask_1)
word_cloud = WordCloud(mask = rgb_array, background_color = 'white', max_words = 1010, repeat = True , colormap ='plasma')
word_cloud.generate(email_body)

plt.imshow(word_cloud, interpolation= 'bilinear')
plt.axis('off')
plt.show()
```
