### $$ MSE = \frac {1}{n} \sum \limits_{i=1} ^ {n} (y_i -\hat y_i)^2 $$
$ \hat y_i = \theta_0 + \theta_1 .x_i $

## $$ MSE = \frac {1}{n} \sum \limits_{i=1} ^ {n} (y_i - (\theta_0 + \theta_1 x_i))^2 $$
Belows are the partial derivatives with respect to.... $\theta_0 $  &  $ \theta_1$

### $$ \frac {\partial MSE} {\partial \theta_0}  = -\frac {2}{n} \sum \limits_{i=1} ^ {n} (y_i - \theta_0 - \theta_1 x_i) $$

### $$ \frac {\partial MSE} {\partial \theta_1}  = -\frac {2}{n} \sum \limits_{i=1} ^ {n} (y_i - \theta_0 - \theta_1 x_i)(x_i) $$

```python
def grad(x , y , thetas):
    n = y.size
    theta0_slope = (-2/n)* sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n)* sum((y - thetas[0] - thetas[1]*x)*x)
    return np.array([theta0_slope[0] , theta1_slope[0]])
```

```python
thetas = np.array([4.8 ,4.8])                           # any value from the graph
```

```python
thetas[0]
```

```python
thetas[1]
```

```python
grad(x , y , thetas)
```

```python
multiplier = 0.01
thetas = np.array([4.8 , 4.8])
plot_vals = thetas.reshape(1,2)
mse_vals = mse(y, thetas[0]+ thetas[1]*x)

for i in range(1000):    
    thetas = thetas - multiplier * grad(x, y, thetas)    
    mse_vals = np.append(arr=mse_vals, values=mse(y, thetas[0] + thetas[1]*x))    
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1, 2)), axis=0)
```

```python
fig=plt.figure(figsize=(16, 12)).add_subplot( projection='3d')

fig.set_xlabel('Theta_0', fontsize=20)
fig.set_ylabel('Theta_1', fontsize=20)
fig.set_zlabel('MSE - The Cost', fontsize=20)

fig.plot_surface(th_0, th_1, MSE, alpha = 0.4, cmap = cm.cool)
fig.scatter(plot_vals[:,0],plot_vals[:,1], mse_vals, c = 'red' ) 
fig.azim=-45

plt.show()
```
