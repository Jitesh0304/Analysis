# Bar plot

```python
sns.barplot( x = data_1.index , y =data_1['OO'])

plt.xlabel('abc')
plt.ylabel('oup')
```

# Heat map

```python
plt.figure(figsize = (18,6))
sns.heatmap(data = data_1 , annot = True)
```

# Scatter plot

```python
plt.figure(figsize = (6,4))
sns.scatterplot( x = data_1['F9'] , y=data_1['OO'])
```

```python
plt.figure(figsize = (6,4))
sns.scatterplot( x = data_1['F9'] , y=data_1['OO'] , hue = data_1['EV'])
```

# Regplot

```python
plt.figure(figsize = (6,4))
sns.regplot( x=data_1['F9'], y=data_1['OO'])
```

# Lm_plot

```python
sns.lmplot( x = 'EV' , y = 'F9' , hue = 'OO' ,data = data_1)
```

```python
sns.lmplot( x = 'EV' , y = 'F9' ,data = data_1)
```

# Swarm_plot

```python
sns.swarmplot( x = data_1['EV'] , y = data_1['F9'])
```

```python
sns.swarmplot( x = 'EV' , y = 'F9' , data = data_1)
```

```python
data_iris = pd.read_csv('iris.csv')
```

```python
data_iris.head()
```

# Hist_plot

```python
sns.histplot(data_iris['Sepal Width (cm)'])
```

```python
sns.histplot(data = data_iris , x = 'Petal Length (cm)' , hue = 'Species')

plt.title('Histogram of petal length by species')
```

# kde-plot

```python
sns.kdeplot(data = data_iris['Sepal Width (cm)'])
```

```python
sns.kdeplot(data = data_iris['Sepal Width (cm)'], shade = True)
```

```python
sns.kdeplot(data = data_iris , x ='Sepal Width (cm)',hue = 'Species' , shade = True)

plt.title('Distribution of Sepal Width, by width')
```

# joint-plot

```python
sns.jointplot(x=data_iris['Sepal Length (cm)'] , y =data_iris['Sepal Width (cm)'], kind = 'kde')
```
